```python
1. import random

def pull_bottle():
    # 90% chance of 10 oz, 10% chance of 9 oz
    return 10 if random.random() < 0.9 else 9
```

```python
2. A bottle is:
10 oz with probability 0.9
9 oz with probability 0.1

E(X)=(10)(0.9)+(9)(0.1)=9.9 ounces
Expected weight: 9.9 oz
```

```python
3. weights = [pull_bottle() for _ in range(1000)]
simulated_average = sum(weights) / len(weights)
simulated_average
```

```python
5. def pulls_until_short():
    count = 0
    while True:
        count += 1
        if pull_bottle() == 9:
            return count

trials = [pulls_until_short() for _ in range(1000)]
sum(trials) / len(trials)
```

```python
7.

def guarantee_pink(stack):
    temp = []
    while True:
        plate = stack.pop()
        if plate == "pink":
            # return rejected plates
            while temp:
                stack.append(temp.pop())
            return plate
        else:
            temp.append(plate)
```

## 1 Maximize the function:

$$ f(x) = a\ln\left(\frac a x\right) -a +x$$

```python
import sympy

x = sympy.Symbol("x")
a = sympy.Symbol("a")

f = a * sympy.log(a/x) - a + x

sympy.solve(f.diff(x),x)
```
